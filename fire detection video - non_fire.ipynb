```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('D:/FCI/graduation project/smart city/Machine Learning/Fire_Detection_Model.h5')
```

```python
import cv2
import numpy as np
```

```python
video = cv2.VideoCapture("D:/FCI/graduation project/smart city/Machine Learning/fire and non fire images and videos/non fire video 1.mp4")
```

```python
def preprocessing_image(frame):
    if frame is None:
        print("Error: Could not load image")
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR) #convert
    frame = cv2.resize(frame, (196,196))  # resize
    frame = frame / 255 #scale
    return frame
```

```python
X = []
while (video.isOpened()): # while this loop is running, we will extract the frames from the video
    ret, frame = video.read() #ret is a boolean value that indicates whether we are getting frames or not #frame is the variable that stores the frames
    #Grabs, decodes and returns the next video frame.
    if ret==False: # if the video ends, go out of the loop
        print("video ended")
        break
    
    frame = cv2.resize(frame, (1200,700))
    cv2.imshow('Frame',frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
    img = preprocessing_image(frame)
    X.append(img)
    
video.release()
cv2.destroyAllWindows()
```

```python
X = np.array(X)
```

```python
X.shape
```

```python
y_pred = model.predict(X)
```

```python
y_pred = y_pred > 0.5
```
